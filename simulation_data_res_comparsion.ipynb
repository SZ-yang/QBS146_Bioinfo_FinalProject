```python
SIM_DIR = "./sim_datasets"
RES_DIR = "./results/sim"

MG_DIR = os.path.join(RES_DIR, "moran_geary")
SPARKX_DIR = os.path.join(RES_DIR, "SparkX")
SOMDE_DIR = os.path.join(RES_DIR, "somDE")

OUT_DIR = "./results/sim/combined_eval"
os.makedirs(OUT_DIR, exist_ok=True)

# Difficulty -> sim path
sim_paths = {
    "easy": os.path.join(SIM_DIR, "sim_easy.h5ad"),
    "medium": os.path.join(SIM_DIR, "sim_medium.h5ad"),
    "hard": os.path.join(SIM_DIR, "sim_hard.h5ad"),
}
```

```python
sims = {k: ad.read_h5ad(v) for k, v in sim_paths.items()}

truth = {}
for diff, sim in sims.items():
    truth_df = sim.var[["is_spatial", "pattern_type"]].copy()
    truth_df.index.name = "gene"
    truth[diff] = truth_df

    print(diff, sim.shape, truth_df["pattern_type"].value_counts().to_dict())
```
